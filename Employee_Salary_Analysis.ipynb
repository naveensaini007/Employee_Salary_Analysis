```python
import pandas as pd
salary = pd.read_csv("/content/employee_salary_dataset.csv")
salary.head()
```

```python
sorted_column = salary.sort_values(by='Department')
sorted_column
```

```python
Average_salary = salary.groupby('Department')['Monthly_Salary'].mean()
Average_salary
```

```python

highest_paid_employee = salary.loc[salary['Monthly_Salary'].idxmax()]

lowest_paid_employee = salary.loc[salary['Monthly_Salary'].idxmin()]

print("Highest Paid Employee:")
display(highest_paid_employee)

print("\nLowest Paid Employee:")
display(lowest_paid_employee)
```

```python
average_salary_by_gender = salary.groupby('Gender')['Monthly_Salary'].mean()
display(average_salary_by_gender)
```
